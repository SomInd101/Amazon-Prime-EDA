````python
# Year wise total content uploading
year_wise_upload = df.groupby(['release_year']).agg({"release_year":"count"}).rename(columns={'release_year': 'content_count'})


# Year and category wise total content uploading
#If we use **"df_working_set"** then the content count will be much higher then the actual because the **"genres"** column data is already sorted.
year_wise_contentCategory = df.groupby(['release_year','type']).agg({"type":"count"}).rename(columns={'type': 'category_count'})

# Year and genres wise total content uploading
year_wise_genres = df_working_set.groupby(['release_year','genres']).agg({"genres":"count"}).rename(columns={'genres': 'genres_count'})

# Year, category and genres wise total content uploading
year_wise_category_genres = df_working_set.groupby(['release_year','type','genres']).agg({"type":"count","genres":"count"}).rename(columns={'genres': 'genres_count','type': 'category_count'})
````
